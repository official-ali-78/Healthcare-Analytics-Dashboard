### Age Analysis

```python
bins = [0,18,35,50,65,100]

labels = ['0-18','19-35','36-50','51-65','65+']

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels
)

df.groupby('Age_Group')['Billing Amount'].mean()
```
